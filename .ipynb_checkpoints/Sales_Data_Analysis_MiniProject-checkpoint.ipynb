## Task 4:
#### Find out the monthy revenue and analyze the findings

a.) Create a column with Year and month field only give a suitable name to it.

```python
order_date_year = data["Order Date"].dt.year
data['Year'] = pd.DataFrame(order_date_year)
order_date_month = data["Order Date"].dt.month
data['Month'] = pd.DataFrame(order_date_month)

monthly_data = data['Order Date'].dt.strftime('%Y-%m')
monthly_data
```

b.) Create a separate dataset from the data which will have two columns, one which is created in step 'a' of this task and second the monthly sales/revenue

```python
monthly_sales = data.groupby(['Year','Month'])['Sales'].sum()
newdata = pd.DataFrame(monthly_sales)
print(newdata)

newdata1 = data.groupby(['Year','Month'])['Profit'].sum()
newdata1 = pd.DataFrame(newdata1)
print(newdata1)
```

```python
result_month = newdata.merge(newdata1, left_index=True, right_index=True)
result_month.reset_index(inplace=True)
result_month['Date'] = pd.to_datetime(result_month[['Year','Month']].assign(DAY=1))
tempmonth = result_month
result_month = result_month.drop(columns=["Year","Month"]) 
result_month
```

c.) Plot this dataset

```python
result_month.plot(x='Date',y=['Sales','Profit'])
```

## Task 5:
#### Find out the yearly revenue and analyze the findings

a.) Create a column with Year field only give a suitable name to it

```python
yearly_data = data['Order Date'].dt.strftime('%Y')
yearly_data
```

b.) Create a separate dataset from the data which will have two columns,
one which is created in step a of this task and second the yearly sales/revenue


```python
yearly_sales = data.groupby(['Year'])['Sales'].sum()
newdata = pd.DataFrame(yearly_sales)
print(newdata)

newdata1 = data.groupby(['Year'])['Profit'].sum()
newdata1 = pd.DataFrame(newdata1)
print(newdata1)

result_year = newdata.merge(newdata1, left_index=True, right_index=True)
result_year.reset_index(inplace=True)
result_year
```

c.) Plot this dataset

```python
result_year.plot.bar(x='Year',y=['Sales','Profit'])
```

d.) Describe your analysis for the yearly revenue observations.

```python
result_year['% Diff Sales'] = result_year.groupby('Year')['Sales'].pct_change()
result_year['% Diff Profit'] = result_year.groupby('Year')['Profit'].pct_change()
result_year
```

### Inferences:

1.) The Sales Revenue and profit have been on an increase with execption of 2015, where the sales declined.

2.) The Sales for the Year 2015 had a lower turnout by a differnece of  13714.9891(0.02833%)  as compared to 2014.Sales of 2016 increased by a difference of 138673.089(0.294715%).

3.) While the sales for 2015 were comparatively lower to 2014, the profit still increased in 2015 by 0.243715%.
